```python
import matplotlib.pyplot as plt
```

```python
import numpy as  np
x=np.arange(0,10)
```

```python
x
```

```python
y=x**2
```

```python
y
```

```python
plt.plot(x,y,'r*-')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('square plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'g*-')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,0.5,1])
axes2=fig.add_axes([0.5,0,0.5,1])
axes1.plot(x,y,'r')
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_title('map')

axes2.plot(y,x,'b')
axes2.set_xlabel('x axis')
axes2.set_ylabel('y axis')
axes2.set_title('map')
```

```python
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
ax2=fig1.add_axes([0,0,1,1])
ax1.plot(x,x**2,label='x**2')
ax2.plot(x,x**3,label='x**3')
ax1.legend(loc=2)
```

```python
fig2=plt.figure()
ax1=fig2.add_axes([0,0,1,1])
ax2=fig2.add_axes([0,0,1,1])
ax1.plot(x,y,label='x**2')
ax2.plot(y,x,label='x**2 reverse')
ax1.legend(loc=1)
ax1.set_xlabel('square of nos')
ax1.set_ylabel('nos')
ax1.set_title('map')
```

```python
fig3=plt.figure()
ax1=fig3.add_axes([0,0,1,1])
ax2=fig3.add_axes([0.3,0.5,0.4,0.4])
ax1.plot(x,y)
ax2.plot(y,x)
ax1.set_xlabel('square of nos')
ax1.set_ylabel('nos')
ax2.set_xlabel('square of nos')
ax2.set_ylabel('nos')
ax1.set_title('map')
```

```python
fig,ax=plt.subplots(1,2)
ax[0].plot(x,y)
ax[1].plot(x,y)
ax[1].set_xlim([2,5])
ax[1].set_ylim([0,60])

```
